```python
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib 
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize']=(12,8) #adjusts the configuration of the plots we will create

df = pd.read_csv('/Users/gacau/Documents/movie ind/movies.csv') #read in the data

df.head()
```

```python
#drop all rows that have missing data
df = df.dropna()
```

```python
#cleaning data
#check for missing data within columns

for col in df.columns:
    pctMissing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pctMissing))
```

```python
#data type
df.dtypes
```

```python
#change data type of columns - get rid of the .0 on budget and gross
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
df.head()
```

```python
#here for reference, my data matches years so I dont have to
#create new column(year) to math release data year 
#df['yearcorrect'] = df['released'].astype(str).str[4:11:]
#del df['yearcorrect'] #delete column
df.head()
```

```python
#order data by gross revenue
df = df.sort_values(by=['gross'], inplace= False, ascending=False)
```

```python
#view all the data
pd.set_option('display.max_rows', None)
```

```python
#drop any duplicates 
#df.drop_duplicates() #drop across data
#this shows distinct values
df['company'].drop_duplicates().sort_values(ascending=False).head()
```

```python
df.head()
```

```python
#budget will have a high correlation
plt.scatter(x=df['budget'],y=df['gross'])

plt.title('Budget vs Gross Earnings')

plt.ylabel('Gross Earnings')
plt.xlabel('Budget for Film')

plt.show()

```

```python
df.head()
```

```python
#regression plot the budget vs gross using seaborn 
sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color": "black"}, line_kws={"color": "blue"})
```

```python
#looking at correlation 
#different types of corr - pearson(default), kendall, spearman - beaware of which one you are using
#highest corr between budget and gross
df.corr(method='pearson')
```

```python
correlationMatrix = df.corr(method='pearson')
sns.heatmap(correlationMatrix, annot = True, cmap= 'coolwarm')
plt.title('Correlation Matrix')

plt.ylabel('Movie Features')
plt.xlabel('Movie Features')
plt.show()
```

```python
df.head()
```

```python
#converting object fields to integers
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype ('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized.head()
```

```python
df.head()
```

```python
#every single fields correlation
correlationMatrix = df_numerized.corr(method='pearson')
sns.heatmap(correlationMatrix, annot = True, cmap= 'coolwarm')
plt.title('Correlation Matrix')

plt.ylabel('Movie Features')
plt.xlabel('Movie Features')
plt.show()
```

```python
df_numerized.corr()
```

```python
#show which one has the highest corr using unstacking
correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()
corr_pairs
```

```python
sorted_pairs=corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs)>0.5]

high_corr
```
